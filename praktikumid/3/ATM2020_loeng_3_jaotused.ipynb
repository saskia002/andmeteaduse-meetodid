```python
np.random.seed(0)
for i in range(1,6):
    n = poisson(6).rvs(1)[0]
    x = uniform(loc=-3,scale=6).rvs(n)
    fig = plt.figure(figsize=(8,0.5)) #facecolor='white')
    ax = plt.axes(frameon=False)
    ax.axes.get_yaxis().set_visible(False)
    plt.xticks(fontsize=20)
    plt.xlim((-3.2,3.2))
    plt.ylim((-0.01,0.1))
    xmin, xmax = ax.get_xaxis().get_view_interval()
    ymin, ymax = ax.get_yaxis().get_view_interval()
    ax.add_artist(plt.Line2D((xmin, xmax), (ymin, ymin), color='black', linewidth=2))
    ax.plot(x,0*x+0.06,'o',color='red')
    plt.gcf().subplots_adjust(bottom=0.7)
    plt.savefig('loeng_3_joonis_9_%i_poissoni_protsess.png'%i)
    plt.show()
```

```python
x = uniform(0,1).rvs(1000100)
y = 0*x[:1000000]
show = [1,2,4,8,16,32,64,128]
show = [1,2,3,4,5,6,7,100]
fig, axs = plt.subplots(2,4,sharey='all')
row = 0
col = 0
for i in range(1,101):
    y = y + x[i:(1000000+i)]
    if i in show:
        h = np.histogram(y/i,100)
        axs[row,col].plot(h[1][:100],h[0]/max(h[0]),'-')
        axs[row,col].text(h[1][50],0.5,'n=%i'%i,fontsize=20,horizontalalignment='center')
        col += 1
        if col>3:
            col = 0
            row += 1
plt.savefig('loeng_3_joonis_17_sum_uniform.png')
plt.show()
```
